```python
import pandas as pd
import matplotlib.pyplot as plt

# Laden der Daten aus den CSV-Dateien
train_df = pd.read_csv('train.csv')
ideal_df = pd.read_csv('ideal.csv')
test_df = pd.read_csv('test.csv')

# Visualisierung der Trainingsdaten
plt.figure(figsize=(10, 6))
for i in range(1, 5):
    plt.scatter(train_df['x'], train_df[f'y{i}'], label=f'y{i}')
plt.title('Visualisierung der Trainingsdaten')
plt.xlabel('X-Wert')
plt.ylabel('Y-Werte')
plt.legend()
plt.show()

# Visualisierung der idealen Funktionen
plt.figure(figsize=(15, 9))
for i in range(1, 51):
    plt.scatter(ideal_df['x'], ideal_df[f'y{i}'], label=f'y{i}')
plt.title('Visualisierung der idealen Funktionen')
plt.xlabel('X-Wert')
plt.ylabel('Y-Werte')
plt.legend()
plt.show()

# Visualisierung der Testdaten
plt.figure(figsize=(10, 6))
plt.scatter(test_df['x'], test_df['y'], color='red', label='Testdaten')
plt.title('Visualisierung der Testdaten')
plt.xlabel('X-Wert')
plt.ylabel('Y-Wert')
plt.legend()
plt.show()
```
